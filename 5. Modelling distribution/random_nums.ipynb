Generating random numbers
------

Analytic CDFs can be used to generate random numbers with a given distribution function, p = CDF(x). If there is an efficient way to compute
the inverse CDF, we can generate random values with the appropriate distribution by choosing p from a uniform distribution between 0 and 1, then
choosing x = ICDF(p)

For example, the CDF of the exponential distribution is

p = 1 − e^(−λx)

Solving for x yields:

x = − log(1 − p)/λ

```python
import numpy as np

p = np.random.rand(25)
def invCDF(p,lam):
    return -np.log(1-p)/lam
lam = 2
values = [invCDF(x,lam) for x in p]
```

```python
def eval_cdf(value,series):
    count =0
    for x in series:
        if x <= value:
            count += 1
    return count/len(series)

def compute_cdf(series):
    x = sorted(series)
    y = []
    for value in x:
        y.append(eval_cdf(value,x))
    return x,y

def exponential(val,lam):
    return 1-np.exp(1)**(-lam*val)
```

```python
# random values generated from inverse CDF
x,y = compute_cdf(values)

# model plot for exponential distribution
y1 = [exponential(val,lam) for val in x]
import matplotlib.pyplot as plt
plt.plot(x,y,label='CDF from random nos.')
plt.plot(x,y1,label='Exponential CDF Model')
plt.xlabel('values')
plt.ylabel('CDF')
plt.legend()
plt.show()
```

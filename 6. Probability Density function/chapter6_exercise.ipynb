```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv(os.path.join('..','data','hinc.csv'))
```

```python
df= df[df['income']<250000]
```

```python
fig,(ax1,ax2) = plt.subplots(ncols=2)
fig.set_size_inches(15,5)
ax1.plot(df['income'],df['ps'])
ax1.set_title('Given data')
ax2.plot(np.log10(df['income'].values),df['ps'])
ax2.set_title('Lognormal plot')
plt.plot()
```

To estimate mean and other statistics from these data, we have to make some
assumptions about the lower and upper bounds, and how the values are distributed in each range.

The normal distribution

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# find the probability of the values occuring less than 1.65+mean
# here the mean is 0 

norm.cdf(1.65)

# this will give percentage of the values which are less than the given value i.e = 1.65+mean
```

Percent point function

```python
# we can also draw out the Z-value by feeding the percentile value
norm.ppf(0.95)
```

```python
# plot a pdf based on normal distribution
fig,ax = plt.subplots()

z_values=np.linspace(norm.ppf(0.001,loc=10,scale=1),norm.ppf(0.9999,loc=10,scale=1),100)
prob = [norm.pdf(x,loc=10,scale=1) for x in z_values]
ax.plot(z_values,prob)
```
